```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_excel('ccpp.xlsx')
data.head()
```

```python
def linear_regression(x,y):
    n = len(x)
    xbar = x.mean()
    ybar = y.mean()
    a = sum(x*y)
    b = sum(x*x)
    b1 = (a - n*xbar*ybar)/(b - n*xbar*xbar)
    b0 = ybar - b1 * xbar
    return b0,b1
```

```python
def test(b0,b1,x,y):
    pred = b0 + b1*x
    rmse = np.sqrt(sum((pred - y)**2))
    return rmse
```

```python
size = len(data);
train_size = int((7*size)/10)
traindata = data[:train_size]
testdata = data[train_size:]
```

```python
# AT Vs PE

sns.lmplot(x='AT',y='PE',data=traindata,scatter_kws = {'s':2},line_kws={"color":"r","lw":2})
b0,b1 = linear_regression(traindata.AT,traindata.PE)
rmse = test(b0,b1,testdata.AT,testdata.PE)
print(' AT Vs PE : b0 = {} b1 = {} RMSE = {}'.format(round(b0,3),round(b1,3),round(rmse,3)))
```

```python
# V Vs PE

sns.lmplot(x='V',y='PE',data=traindata,scatter_kws = {'s':2},line_kws={"color":"r","lw":2})
b0,b1 = linear_regression(traindata.V,traindata.PE)
rmse = test(b0,b1,testdata.V,testdata.PE)
print(' V Vs PE : b0 = {} b1 = {} RMSE = {}'.format(round(b0,3),round(b1,3),round(rmse,3)))
```
